```python
from modules.Extractor import *
from modules.Transformer import *
from modules.Loader import *
```

```python
# ETL Pipeline to generate the data of customers who bought Airpods just after buying Iphone
class ProcessIphoneToAirpodConversions:
    def __init__(self):
        pass

    def runner(self):

        # Step 1: Extracting the required data
        input_df = DataExtractor().extract()

        # Step 2: Implementing the Transformation logic
        iphone_to_airpods_buyers = IphoneToAirpodsConversions().transform(input_df)

        # Step 3: Loading the data to the sink
        IphoneToAirpodsConversionsLoader(iphone_to_airpods_buyers).sink()
```

```python
# Customers who bought only Iphone and Airpods
class ProcessIphoneAndAirpodBuyers:
    def __init__(self):
        pass

    def runner(self):

        # Step 1: Extracting the required data
        input_df = DataExtractor().extract()

        # Step 2: Implementing the Transformation logic
        iphone_and_airpods_buyers = ExclusiveIphoneAndAirpodsBuyers().transform(input_df)

        # Step 3: Loading the data
        ExclusiveIphoneAirpodsLoader(iphone_and_airpods_buyers).sink()
```

```python
# To find the top 3 products in each category based on the revenue generated
class ProcessCategoryProductLeaderboard:
    def __init__(self):
        pass

    def runner(self):

        # Step 1: Extracting the required data
        input_df = DataExtractor().extract()

        # Step 2: Implementing the Transformation logic
        iphone_and_airpods_buyers = CategoryProductLeaderboard().transform(input_df)

        # Step 3: Loading the data
        CategoryProductLeaderboardLoader(iphone_and_airpods_buyers).sink()
```

```python
class WorkFlowRunner():
    def __init__(self, name):
        self.name = name
    
    def runner(self):
        
        if self.name == "ProcessIphoneToAirpodConversions":
            return ProcessIphoneToAirpodConversions().runner()
        elif self.name == "ProcessIphoneAndAirpodBuyers":
            return ProcessIphoneAndAirpodBuyers().runner()
        else:
            raise ValueError(f"Not Implemented for {self.name}")
```
